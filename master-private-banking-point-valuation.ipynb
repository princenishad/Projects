# **TARGET 1 : UNSPECIFIED TIME PERIOD**

**Objective : To get a broader view and overwhelmingly optimistic target for a scrip**

Target_one = (Ind PE + CMP/FCF ) * CMP

```python
df['Target_one'] = round(((df['Ind PE'] - df['PE']) + df['CMP / FCF'])*df['LTP'],2)
df
```

# **Target 2 : Foresing only Micro Economic factors**

**Objective : To take only those factors which affect scrips individually and not across the sector**

**Adjusted Return to Shareholder : ROE/PB**

```python
df['Adjusted_return_to_shareholder'] = round(df['ROE']/df['PB'],2)
```

**Calculating Flexibility**

```python
promotor_distribution = df['PROMOTER'].quantile([0.25,0.50,0.75])
promotor_distribution
```

```python
flexibility_index = []
for holding in df['PROMOTER']:
    if holding <= 0.00:
        flexibility_index.append(int(1))
    elif 0.00 < holding <= 8.19:
        flexibility_index.append(int(2))
    elif 8.19 < holding <= 39.93:
        flexibility_index.append(int(3))
    elif 39.93 < holding <= 59.00:
        flexibility_index.append(int(4))
    else:
        flexibility_index.append(int(10))
        
```

```python
df['Flexibilty'] = flexibility_index
```

**CR = CMP(1 + r/n)^nt**

```python
# pre defined varaibles 
# n = 12
# t= 1
df['Target_two'] = round(df['LTP']*((1+(df['3mth_return_%']+df['Adjusted_return_to_shareholder'])/(4*100))**(4*1)),2)
```

```python
df
```

```python
operating_profit = df.iloc[:,16:17]
operating_profit.columns = ['Operating_Profit_growth']
```

# **Target 3 : Foresing all the Macro Economic Indicators**

```python
## Pre defined variables
count_of_rate_hikes = 2 
N = 3
inflation_rate = 0.067 ## RBI DATA
sector_growth_rate = 0.08 ## Mordor Intelligence DATA
lambda_ = 0.80

df['Target_three'] = round(((operating_profit['Operating_Profit_growth']/100 + 1)**(N - count_of_rate_hikes))*(df['LTP']*(1+(inflation_rate)*(lambda_)+sector_growth_rate)),2)
df
```

**NOW THE THING IS THAT ASSET QUALITY IS ANALYZED WITH MULTIPLE FACTORS KEEPING IN MIND SUCH AS PROVISIONS AND CONTANGENCIES WHICH REQUIRES MAN-MADE INTERVENTION.**

**THE CONSTANT ALLOCATION WILL BE DONE BY KEEPING IN MIND THE SUMMED UP EXPECTATION OF EACH SCRIP**

**TARGET_ONE EXPECTATION : 0.025,**
**TARGET_TWO EXPECTATION : 0.125,**
**TARGET_THREE EXPECTATION : 0.85,**

****

```python
df['Final_Target'] = round(df['Target_one']*(0.025) + df['Target_two']*(0.125) + df['Target_three']*(0.85), 2)
```

```python
df['Expectation%'] = round((df['Final_Target'] - df['LTP'])*100/df['LTP'],2)
```

```python
final_df = df.sort_values('Expectation%', ascending=False)
final_df
```

```python
final_df = final_df.reset_index()
final_df.drop(['index'],axis=1,inplace=True)
final_df
```

**NOW COMES THE CONSTANT ALLOCATION WHICH IS NOTHING BUT DISTRIBUTION OF TOP 5 FLEXIBILITY INDEX VALUE**

```python
series_A = final_df.iloc[0:5,:]
series_A
```

```python
series_A['Allocation'] = series_A['Flexibilty']/sum(series_A['Flexibilty'])
```

```python
series_A
```

**Let us start with initial allocation of INR 10K**

```python
series_A['Corpus'] = 10000
series_A['Invested_Amount'] = round(series_A['Corpus']*series_A['Allocation'],2)
series_A['Allocated-Shares'] = round(series_A['Invested_Amount']/series_A['LTP'],0)
```

```python
series_A
```
